```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------
# Load your dataset here
# ----------------------
# Example: df = pd.read_csv('your_data.csv')
# Make sure columns include: test (0/1), converted (0/1), mode_impr_day (1â€“7)

# Step 1: Clean column names if needed
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Step 2: Compute conversion counts
grouped = df.groupby('test')['converted'].agg(['sum', 'count']).rename(columns={'sum': 'conversions', 'count': 'total'})
test_group = grouped.loc[1]
control_group = grouped.loc[0]

# Q3: Chi-squared test
conversion_data = [
    [control_group['conversions'], control_group['total'] - control_group['conversions']],
    [test_group['conversions'], test_group['total'] - test_group['conversions']]
]

chi2_stat, p_val, dof, expected = chi2_contingency(conversion_data)

# Print Q3 results
print("\n--- Q3: Advertising Campaign Effectiveness ---")
print(f"Test Group: {int(test_group['conversions'])} conversions out of {int(test_group['total'])} users â†’ {test_group['conversions'] / test_group['total']:.2%}")
print(f"Control Group: {int(control_group['conversions'])} conversions out of {int(control_group['total'])} users â†’ {control_group['conversions'] / control_group['total']:.2%}")
print(f"Difference: {(test_group['conversions'] / test_group['total'] - control_group['conversions'] / control_group['total']):.2%}")
print("95% CI: Approximate from MedCalc output: 0.578% to 0.927%")
print(f"Chi-squared Statistic: {chi2_stat:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p_val:.5f}")

# ------------------------------------------------------
# Q4: Optimal Days for Advertising - Conversion by Weekday
# ------------------------------------------------------

# Map numeric day to weekday name
day_map = {
    1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday",
    5: "Friday", 6: "Saturday", 7: "Sunday"
}
df['weekday'] = df['mode_impr_day'].map(day_map)

# Compute average conversion rate by test group and weekday
calendar_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
conversion_by_day = (
    df.groupby(['test', 'weekday'])['converted']
    .mean()
    .reset_index()
    .pivot(index='weekday', columns='test', values='converted')
    .reindex(calendar_order)
)

conversion_by_day.columns = ['Control (0)', 'Exposed (1)']
conversion_by_day.reset_index(inplace=True)

# Print Q4 table
print("\n--- Q4: Conversion Rate by Day of Week (Test vs Control) ---")
print(conversion_by_day)

# Plot
plt.figure(figsize=(10, 5))
melted = pd.melt(conversion_by_day, id_vars='weekday', var_name='Group', value_name='Conversion Rate')
sns.barplot(data=melted, x='weekday', y='Conversion Rate', hue='Group', palette='viridis')
plt.title("Conversion Rate by Day of the Week (Test vs Control)")
plt.xlabel("Day of Week")
plt.ylabel("Conversion Rate")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, norm
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------
# Load your dataset here
# ----------------------
# Example: df = pd.read_csv('your_data.csv')
# Ensure columns: test (0/1), converted (0/1), mode_impr_day (1â€“7)

# Step 1: Clean column names if needed
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Step 2: Compute conversion counts by group
grouped = df.groupby('test')['converted'].agg(['sum', 'count']).rename(columns={'sum': 'conversions', 'count': 'total'})
test_group = grouped.loc[1]
control_group = grouped.loc[0]

# Conversion rates
p1 = test_group['conversions'] / test_group['total']
p2 = control_group['conversions'] / control_group['total']
diff = p1 - p2

# Standard error of difference in proportions (Wald method)
se_diff = np.sqrt(p1 * (1 - p1) / test_group['total'] + p2 * (1 - p2) / control_group['total'])

# 95% confidence interval (Z = 1.96 for 95%)
z = norm.ppf(0.975)
ci_lower = diff - z * se_diff
ci_upper = diff + z * se_diff

# Chi-squared test
conversion_data = [
    [control_group['conversions'], control_group['total'] - control_group['conversions']],
    [test_group['conversions'], test_group['total'] - test_group['conversions']]
]
chi2_stat, p_val, dof, expected = chi2_contingency(conversion_data)

# Print Q3 results
print("\n--- Q3: Advertising Campaign Effectiveness ---")
print(f"Test Group: {int(test_group['conversions'])} conversions out of {int(test_group['total'])} users â†’ {p1:.2%}")
print(f"Control Group: {int(control_group['conversions'])} conversions out of {int(control_group['total'])} users â†’ {p2:.2%}")
print(f"Difference in conversion: {diff:.2%}")
print(f"95% Confidence Interval for difference: {ci_lower:.4%} to {ci_upper:.4%}")
print(f"Chi-squared Statistic: {chi2_stat:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p_val:.5f}")

# ------------------------------------------------------
# Q4: Optimal Days for Advertising - Conversion by Weekday
# ------------------------------------------------------

# Map numeric day to weekday name
day_map = {
    1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday",
    5: "Friday", 6: "Saturday", 7: "Sunday"
}
df['weekday'] = df['mode_impr_day'].map(day_map)

# Compute conversion rate by group and weekday
calendar_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
conversion_by_day = (
    df.groupby(['test', 'weekday'])['converted']
    .mean()
    .reset_index()
    .pivot(index='weekday', columns='test', values='converted')
    .reindex(calendar_order)
)

# Rename for clarity
conversion_by_day.columns = ['Control (0)', 'Exposed (1)']
conversion_by_day.reset_index(inplace=True)

# Print Q4 conversion table
print("\n--- Q4: Conversion Rate by Day of Week (Test vs Control) ---")
print(conversion_by_day)

# Plotting grouped bar chart
plt.figure(figsize=(10, 5))
melted = pd.melt(conversion_by_day, id_vars='weekday', var_name='Group', value_name='Conversion Rate')
sns.barplot(data=melted, x='weekday', y='Conversion Rate', hue='Group', palette='viridis')
plt.title("Conversion Rate by Day of the Week (Test vs Control)")
plt.xlabel("Day of Week")
plt.ylabel("Conversion Rate")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
